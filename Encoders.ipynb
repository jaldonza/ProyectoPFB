```python
import pandas as pd
import pickle
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sqlalchemy import create_engine

DB_PATH = "C:/Users/Khalid/Desktop/PFB/sp500.db"  # Ruta de tu base de datos
engine = create_engine(f'sqlite:///{DB_PATH}')
query = "SELECT id_empresa, simbolo, nombre_empresa, sector, industria FROM empresas_sp500"
empresas_df = pd.read_sql(query, engine)

      # 1. Label Encoding para 'simbolo'
label_encoder_simbolo = LabelEncoder()
empresas_df['simbolo_encoded'] = label_encoder_simbolo.fit_transform(empresas_df['simbolo'])

     # Guardar el encoder para 'simbolo' en un archivo pickle
with open("simbolo_encoder.pkl", "wb") as f:
    pickle.dump(label_encoder_simbolo, f)

     # One-Hot Encoding para 'sector'
one_hot_encoder_sector = OneHotEncoder(sparse_output=False)  
sector_encoded = one_hot_encoder_sector.fit_transform(empresas_df[['sector']])

     # Convertir los resultados de One-Hot Encoding a un DataFrame
sector_encoded_df = pd.DataFrame(sector_encoded, columns=one_hot_encoder_sector.get_feature_names_out(['sector']))

     # AÃ±adir las columnas de One-Hot Encoding al DataFrame original
empresas_df = pd.concat([empresas_df, sector_encoded_df], axis=1)

     # Guardar el encoder para 'sector' en un archivo pickle
with open("sector_encoder.pkl", "wb") as f:
    pickle.dump(one_hot_encoder_sector, f)

      # One-Hot Encoding para 'industria'
one_hot_encoder_industria = OneHotEncoder(sparse_output=False) 
industria_encoded = one_hot_encoder_industria.fit_transform(empresas_df[['industria']])

      # Convertir los resultados de One-Hot Encoding a un DataFrame
industria_encoded_df = pd.DataFrame(industria_encoded, columns=one_hot_encoder_industria.get_feature_names_out(['industria']))

      # AÃ±adir las columnas de One-Hot Encoding al DataFrame original
empresas_df = pd.concat([empresas_df, industria_encoded_df], axis=1)

      # Guardar el encoder para 'industria' en un archivo pickle
with open("industria_encoder.pkl", "wb") as f:
    pickle.dump(one_hot_encoder_industria, f)


empresas_df.head(5)
```
