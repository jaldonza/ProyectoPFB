```python
tickers = sp500_df['Symbol'].tolist()  # Lista de símbolos del S&P 500

# Descargar datos para cada empresa
for ticker in tickers:
    company_data = yf.download(ticker, start='2020-01-01', end='2023-10-01')
    print(f'Datos de {ticker}:')
    print(company_data.head())
```

```python
info_df = pd.DataFrame()
```

for ticker in tickers:
    try:
        company = yf.Ticker(ticker)
        info = company.info
        
        # Verificar si 'info' no está vacío y tiene datos válidos
        if info and isinstance(info, dict) and len(info) > 0:
            # Convertir el diccionario 'info' a un DataFrame temporal con una fila
            temp_df = pd.DataFrame([info])
            
            # Concatenar el DataFrame temporal con el principal
            info_df = pd.concat([info_df, temp_df], ignore_index=True)
        else:
            print(f"Sin datos válidos para {ticker}")

    except Exception as e:
        print(f"Error obteniendo datos de {ticker}: {e}")

# Mostrar las primeras filas del DataFrame con toda la información
