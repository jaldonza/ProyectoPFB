```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as seaborn
import plotly as plt
import plotly.graph_objects as go
```

# Trasformacion de las columnas Date y Timestamp a tipo datetime para la graficación.


```python
df = pd.read_csv('Cotizaciones_historicas_SP_500.csv')

df['Date'] = df['Date'].str.replace(r'([+-]\d{2}:\d{2})', '', regex=True)
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')

print(df[['Date', 'timestamp']].head())


df.to_csv('Cotizaciones_historicas_SP_500.csv', index=False)

print("El archivo CSV ha sido actualizado.")
```

```python
df = pd.read_csv('Cotizaciones_historicas_SP_500.csv')

periodo_inicio = '2023-01-01'
periodo_fin = '2023-12-31'


df_filtrado = df[(df['Date'] >= periodo_inicio) & (df['Date'] <= periodo_fin)]

top_10_empresas = df_filtrado['Symbol'].value_counts().head(10).index.tolist()

colores_pastel = [
   'lightgreen','lightgreen','lightgreen','lightgreen','lightgreen','lightgreen','lightgreen', 'lightgreen', 'lightgreen', 'lightgreen',]


for i, empresa in enumerate(top_10_empresas):
    datos_accion = df_filtrado[df_filtrado['Symbol'] == empresa]
    company_name = datos_accion['Company'].iloc[0]
    
    
    fig = go.Figure(
        go.Box(
            y=datos_accion['Close'],
            name=f'{empresa} - {company_name}',
            boxmean='sd', 
            line=dict(width=2, color='black'),
            marker=dict(color=colores_pastel[i]),  
            boxpoints='all',))
    
    fig.update_layout(
        title=f'Distribución de Precios de Cierre de {empresa} - {company_name}',
        xaxis_title='Precio de Cierre',
        yaxis_title='Frecuencia',
        template='plotly',
        height=600,  
        showlegend=False)
    
    fig.show()
```
